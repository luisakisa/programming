1. Загрузите выборку Wine

```python
import pandas as pd

headers = [
    'quality',
    'Alcohol',
    'Malic acid',
    'Ash',
    'Alcalinity of ash',
    'Magnesium',
    'Total phenols',
    'Flavanoids',
    'Nonflavanoid phenols',
    'Proanthocyanins',
    'Color intensity',
    'Hue',
    'OD280/OD315 of diluted wines',
    'Proline',
]
df = pd.read_csv('wine.data', sep=',', names=headers)
df
```

2. Извлеките из данных признаки и классы.

```python
X = df.drop('quality', axis=1)
y = df['quality']
```

3. Оценку качества необходимо провести методом кросс-валидации по
5 блокам (5-fold). Создайте генератор разбиений, который перемешивает выборку перед формированием блоков (shuffle=True). Для воспроизводимости результата, создавайте генератор KFold с фиксированным параметром random_state=42. В качестве меры качества используйте долю верных ответов (accuracy).
4. Найдите точность классификации на кросс-валидации для метода
k ближайших соседей (sklearn.neighbors.KNeighborsClassifier), при
k от 1 до 50. При каком k получилось оптимальное качество? Чему
оно равно (число в интервале от 0 до 1)? Данные результаты и
будут ответами на вопросы 1 и 2.

```python
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
import numpy as np


kf = KFold(n_splits=5, shuffle=True, random_state=42)
for train_index, test_index in kf.split(X, y):
    negh = []
    for i in range(1, 51):
        knn = KNeighborsClassifier(n_neighbors=i)
        X_train, X_test = X.loc[train_index, :], X.loc[test_index, :]
        y_train, y_test = y[train_index], y[test_index]
        knn.fit(X_train, y_train)
        negh.append(knn.score(X_test, y_test))

    print(f"n: {np.argmax(negh) + 1}, score: {max(negh)}")
```

5. Произведите масштабирование признаков с помощью функции
sklearn.preprocessing.scale. Снова найдите оптимальное k на кроссвалидации.
6. Какое значение k получилось оптимальным после приведения признаков к одному масштабу? Как изменилось значение качества?
Приведите ответы на вопросы 3 и 4.



```python
from sklearn.preprocessing import scale

X_s = scale(X)
kf = KFold(n_splits=5, shuffle=True, random_state=42)
for train_index, test_index in kf.split(X_s, y):
    negh = []
    for i in range(1, 51):
        knn = KNeighborsClassifier(n_neighbors=i)
        X_train, X_test = X_s[train_index, :], X_s[test_index, :]
        y_train, y_test = y[train_index], y[test_index]
        knn.fit(X_train, y_train)
        negh.append(knn.score(X_test, y_test))
    print(f"n: {np.argmax(negh) + 1}, score: {max(negh)}")
```
