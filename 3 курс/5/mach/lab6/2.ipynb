1. Загрузите данные в DataFrame, используя функцию read_csv

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LassoCV

df = pd.read_csv('house_price.csv')
df
```

2. Сколько строк и столбцов в данных? Есть ли пропуски? Для ответа на
вопросы используйте метод info().


```python
df.info()
```

3. Для выполнения задания в наборе данных необходимо оставить только
числовые признаки

```python
numeric_dtypes = ['int64', 'float64']
numerics = []
for i in df.columns:
    if df[i].dtype in numeric_dtypes:
        numerics.append(i)
df = df[numerics]
df
```

4. Удалите столбец Id и пропущенные значения.

```python
del df['Id']
df = df.dropna()
df
```

5. Разделите набор данных на входные данные X (все столбцы кроме
SalePrice) и ответы y (столбец SalePrice).


```python
X = df.drop('SalePrice', axis = 1)
y = df['SalePrice']
```

6. Разделите данные на обучающую и тестовую выборки

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

7. Обучите модель LassoCV (установите значение гиперпараметра cv
самостоятельно). Оцените качество полученной модели. Посмотрите на
коэффициенты модели. Есть ли коэффициенты равные 0? Что это означает?
Попробуйте их убрать и построить модель заново. Как изменилось качество
полученной модели?

```python
lasso = LassoCV(cv=5)
lasso.fit(X_train, y_train)
print(lasso.score(X_test, y_test))

# находим индексы нулей в коэффициентах
zero_idx = np.where(lasso.coef_ == 0)[0]
# убираем соответствующие столбцы в Х
print(len(X.columns))
X = X.drop(X.columns[zero_idx], axis = 1)
print(len(X.columns))
# обучаем заново
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
lasso.fit(X_train, y_train)
print(lasso.score(X_test, y_test))
lasso.coef_
```

Коэф = 0 не влияют на модель. Если их убрать, то качество никак не меняется

8. Попробуйте использовать L2-регуляризацию, т.е. обучите модель RidgeCV.
Сравните полученный результат LassoCV и RidgeCV.

```python
from sklearn.linear_model import RidgeCV

XX = df.drop('SalePrice', axis = 1)
yy = df['SalePrice']
XX_train, XX_test, yy_train, yy_test = train_test_split(XX, yy, test_size=0.3, random_state=42)

ridge = RidgeCV(cv=5)
ridge.fit(XX_train, yy_train)
print(ridge.score(XX_test, yy_test))
print(f"lasso без нулевых коэф.: {lasso.score(X_test, y_test)}\nridge как есть: {ridge.score(XX_test, yy_test)}")
```
